```python
import pandas as pd
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
filename = r'C:\Users\jatto\downloads\projeto massa\perfil_eleitor_secao_2020_SP.csv'                                        
df = pd.read_csv(filename, delimiter=';', 
                 usecols='DS_GENERO DS_GRAU_ESCOLARIDADE DS_FAIXA_ETARIA NM_MUNICIPIO'.split(),
                 encoding='ISO-8859-1',
                 error_bad_lines=False)
```

```python
df.shape
```

```python
df.info()
```

```python
df.groupby ('DS_GENERO').size().sort_values()
```

```python
df.groupby ('DS_GENERO').size().sort_values().plot(kind='bar',color='#629c95' , figsize=(8,8)).set_title('Comparação de gêneros dos eleitores')
```

```python
df.query('DS_GENERO == "FEMININO"').groupby('DS_GRAU_ESCOLARIDADE').size().sort_values(ascending=False).head(10).plot(kind='barh', color='#c9a989', figsize=(8,8)).set_title('Escolaridade das eleitoras')
```

```python
colors = ['#ff8a8a', '#fff9a3', '#39adaa', '#dabfff', '#d6cbd6', '#1135ed', '#e80061','#ecff73','#03572a','#595959']
df.query('DS_GENERO == "FEMININO"').groupby('NM_MUNICIPIO').size().sort_values(ascending=False).head(10).plot(kind='pie', autopct='%.2f%%', colors=colors,  figsize=(8, 8)).set_title('Cidades onde se localizam a maioria das eleitoras')
```

```python
df.query('DS_GENERO == "FEMININO"').groupby('DS_FAIXA_ETARIA').size().sort_values(ascending=False).head(10).plot(kind='bar', color='#fffcbd', figsize=(8,8)).set_title('Faixa etária das eleitoras ')
```
